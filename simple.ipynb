```python
import numpy as np
#import matplotlib.pyplot as plt
import ipywidgets as widgets
import bqplot.pyplot as plt

def debt_v_retirement(monthly_income=10000,
                      monthly_expenses=4000,
                      cash=1000,
                      months=160,
                      loan_principal=250000,
                      loan_interest=0.09,
                      monthly_loan_payment=3000,
                      r_interest=0.09):
    global net_worth
    results=[]
    bank_account=[[0,monthly_income-monthly_expenses-monthly_loan_payment]]
    balance=[[0,loan_principal]]

    retirement=[[0,0]]
    r_payment=monthly_income-monthly_expenses-monthly_loan_payment-cash
    net_worth=[[0,bank_account[0][1]+retirement[0][1]-balance[0][1]]]

    for i in range(months):
    
    #student_loans
        if balance[-1][1]>0:
            new_balance=balance[i][1]*(1+(loan_interest/12))-monthly_loan_payment
            balance.append([i+1,new_balance])
        
            #savings=monthly_income-monthly_expenses+bank_account[i][1]-monthly_loan_payment
            
            new_retirement=retirement[i][1]*(1+(r_interest/12))+r_payment
        
        else:
            #savings=monthly_income-monthly_expenses+bank_account[i][1]
            new_retirement=retirement[i][1]*(1+(r_interest/12))+r_payment+monthly_loan_payment
            balance.append([i+1,0.0])
            
        savings=cash+bank_account[i][1]    
        bank_account.append([i+1,savings])
        retirement.append([i+1,new_retirement])
        
#        net=bank_account[-1][1]+retirement[-1][1]-balance[-1][1]
        net_worth.append([i+1,bank_account[i+1][1]+retirement[i+1][1]-balance[i+1][1]])
    plt.clear()
    #plt.style.use('seaborn')
    plt.plot(np.array(bank_account)[:,0],np.array(bank_account)[:,1],labels=["Cash"],colors=['blue'], display_legend=True,background_style={'fill':'#999'})
    plt.plot(np.array(bank_account)[:,0],np.array(retirement)[:,1],labels=["Retirement"],colors=['green'], display_legend=True,background_style={'fill':'#999'})
    plt.plot(np.array(bank_account)[:,0],-np.array(balance)[:,1],labels=["Debt"],colors=['red'], display_legend=True,background_style={'fill':'#999'})
    plt.plot(np.array(bank_account)[:,0],np.array(net_worth)[:,1],labels=["Net worth"],colors=['black'], display_legend=True,background_style={'fill':'#999'})
    #plt.plot()
    #plt.annotate(str(np.around(np.array(net_worth)[-1,1],decimals=2)), xy=(0.05, 0.95), xycoords='axes fraction')
    #plt.legend(loc="upper center",ncol=2)
    #plt.rcParams['figure.figsize'] = [30.0, 16.0]
    #plt.current_figure()
    plt.show()

    #return bank_account,retirement,balance,net_worth
    
r_int_slider = widgets.FloatSlider(value=0.09,
                          min=0.0,
                          max=0.35,
                          step=0.01,
                          description="retirement interest rate",
                          disabled=False,
                          continuous_update=True,
                          orientation='horizontal',
                          readout=True,
                          reaout_format='d')

loan_int_slider = widgets.FloatSlider(value=0.09,
                          min=0.0,
                          max=0.35,
                          step=0.01,
                          description="loan interest rate",
                          disabled=False,
                          continuous_update=True,
                          orientation='horizontal',
                          readout=True,
                          reaout_format='d')

me_slider = widgets.FloatSlider(value=4000,
                          min=0.0,
                          max=10000,
                          step=1.,
                          description="monthly expenses",
                          disabled=False,
                          continuous_update=True,
                          orientation='horizontal',
                          readout=True,
                          reaout_format='d')

mi_slider = widgets.FloatSlider(value=10000,
                          min=0.0,
                          max=100000,
                          step=1.,
                          description="monthly income",
                          disabled=False,
                          continuous_update=True,
                          orientation='horizontal',
                          readout=True,
                          reaout_format='d')

loan_pay_slider = widgets.FloatSlider(value=3000,
                          min=0.0,
                          max=10000,
                          step=1.,
                          description="monthly loan payments",
                          disabled=False,
                          continuous_update=True,
                          orientation='horizontal',
                          readout=True,
                          reaout_format='d')

cash_slider = widgets.FloatSlider(value=1000,
                          min=0.0,
                          max=10000,
                          step=1,
                          description="cash saved",
                          disabled=False,
                          continuous_update=True,
                          orientation='horizontal',
                          readout=True,
                          reaout_format='d')

loan_bal_slider = widgets.FloatSlider(value=250000,
                          min=0.0,
                          max=1000000,
                          step=1,
                          description="initial loan balance",
                          disabled=False,
                          continuous_update=True,
                          orientation='horizontal',
                          readout=True,
                          reaout_format='d')

mon_slider = widgets.IntSlider(value=240,
                          min=0,
                          max=400,
                          step=1,
                          description="months",
                          disabled=False,
                          continuous_update=True,
                          orientation='horizontal',
                          readout=True,
                          reaout_format='d')

test=widgets.interactive(debt_v_retirement,
                   monthly_income=mi_slider,
                   monthly_expenses=me_slider,
                   cash=cash_slider,
                   months=mon_slider,
                   loan_principal=loan_bal_slider,
                   loan_interest=loan_int_slider,
                   monthly_loan_payment=loan_pay_slider,
                   r_interest=r_int_slider)

output=test.children[-1]
test
```
